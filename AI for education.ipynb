## Text preprocessing

No classical NLP preprocessing is applied, such as:

- removing stop words
- stemming
- lemmatization
- manual tokenization
- lowercasing everything
- removing punctuation

Since the RAG pipeline relies on semantic embeddings, the model already handles tokenization and preserves meaning. Modifying the text could reduce important context, especially in tutoring dialogues.

Therefore, the text is kept in its original form.

# Chunking

### Chunking strategy for Agent 1

For Agent 1, retrieval is based only on the question. At the beginning of the interaction, no incorrect student solution exists yet, since Agent 1 must first generate a flawed response.

Therefore, each chunk is indexed using the question only, while fields such as student_incorrect_solution and teacher_described_confusion are kept as metadata and used after retrieval to guide the generation of a realistic mistake.

```python
chunks = []

for _, row in df.iterrows():
    
    chunk = {
        "text": row["question"],

        "metadata": {
            "qid": row["qid"],
            "student_incorrect_solution": row["student_incorrect_solution"],
            "teacher_described_confusion": row["teacher_described_confusion"],
            "ground_truth": row["ground_truth"],
           
        }
    }
    
    chunks.append(chunk)

len(chunks)
```

# Embedding

### Models selection

We use open-source models for both embeddings and generation to keep the pipeline fully local and reproducible.

For embeddings, we use all-MiniLM-L6-v2, a lightweight model that provides efficient semantic similarity while requiring low computational resources. Since the questions are relatively short (max 148 words), the model’s context capacity is sufficient while maintaining fast inference.

We use Qwen2.5-Instruct as the generator (used later for Agent 1 and Agent 2) because instruction-tuned models are optimized to follow prompts and produce structured, step-by-step explanations, which is well suited for RAG and tutoring tasks.

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
texts = [chunk["text"] for chunk in chunks]

embeddings = model.encode(
    texts,
    show_progress_bar=True,
    normalize_embeddings=True
)

len(embeddings), len(embeddings[0])
```

## ChromaDB

```python
import chromadb

# create database (simple version)
client = chromadb.Client()

# create collection
collection = client.create_collection(name="mathdial", configuration={"hnsw": {"space": "cosine"}})

# prepare data
documents = [chunk["text"] for chunk in chunks]
metadatas = [chunk["metadata"] for chunk in chunks]
ids = [str(i) for i in range(len(chunks))]

# store embeddings
collection.add(
    documents=documents,
    embeddings=embeddings,
    metadatas=metadatas,
    ids=ids
)
```

Cosine similarity is used as the distance metric for retrieval.

```python
def retrieve_examples(query, collection, top_k=5):
    
    # convert query into embedding
    query_embedding = model.encode(query).tolist()
    
    # search similar questions
    results = collection.query(
        query_embeddings=[query_embedding],
        
        n_results=top_k
    )
    
    retrieved_questions = results["documents"][0]
    retrieved_metadata = results["metadatas"][0]
    
    examples = []
    
    for i in range(len(retrieved_questions)):
        
        example = {
            "question": retrieved_questions[i],
            "qid": retrieved_metadata[i]["qid"],
            "student_incorrect_solution": retrieved_metadata[i]["student_incorrect_solution"],
            "teacher_described_confusion": retrieved_metadata[i]["teacher_described_confusion"],
            "ground_truth": retrieved_metadata[i]["ground_truth"],
            
        }
        
        examples.append(example)
    
    return examples
```

```python
query = "There are 5 apples and 10 carrots in the basket."

example = retrieve_examples(query, collection, top_k=2)

example
```

```python
query = "Sarah has 24 eggs. She gives 1/3 of them to her friend. How many eggs does Sarah have left?"

example = retrieve_examples(query, collection, top_k=2)

example
```

## Generator (Agent 1)

```python
import torch
print(torch.cuda.is_available())
print(torch.cuda.get_device_name(0))
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
# Pytorch software library used to run neural networks.
import torch

model_name = "Qwen/Qwen2.5-1.5B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    # 1.5B parameters × 2 bytes ≈ 3 GB VRAM
    torch_dtype=torch.float16,
    # automatically uses GPU if available, otherwise CPU
    device_map="cuda",
    
    
)
```

```python
prompt = "Nancy is filling an aquarium for her fish. She fills it halfway and goes to answer the door. While she's gone, her cat knocks the aquarium over and spills half the water in it. Then Nancy comes back and triples the amount of water in the aquarium. If the aquarium is 4 feet long, 6 feet wide, and 3 feet high, how many cubic feet of water are in the aquarium?"
messages = [

    {"role": "system", "content": """You are a helpful math tutor for 7th grade students.
     Write all mathematical expressions using LaTeX.
     Always enclose any display-style mathematical calculation or equation using double dollar signs, like this: $$...$$. 
     
     
    Respond in JSON format with:
    {{
        
    "is_math_problem": true/false,
    "explanation": "short explanation"

    }}
    
    return ONLY valid JSON, no additional text or markdown formatting"""},
    {"role": "user", "content": prompt}
]
text = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True
)
model_inputs = tokenizer([text], return_tensors="pt").to(model.device)

generated_ids = model.generate(
    **model_inputs,
    max_new_tokens=200
)
generated_ids = [
    output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
]

response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]

print(response)
```

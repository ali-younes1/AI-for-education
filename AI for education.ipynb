## Text preprocessing

No classical NLP preprocessing is applied, such as:

- removing stop words
- stemming
- lemmatization
- manual tokenization
- lowercasing everything
- removing punctuation

Since the RAG pipeline relies on semantic embeddings, the model already handles tokenization and preserves meaning. Modifying the text could reduce important context, especially in tutoring dialogues.

Therefore, the text is kept in its original form.

# Chunking

### Chunking strategy for Agent 1

For Agent 1, retrieval is based only on the question. At the beginning of the interaction, no incorrect student solution exists yet, since Agent 1 must first generate a flawed response.

Therefore, each chunk is indexed using the question only, while fields such as student_incorrect_solution and teacher_described_confusion are kept as metadata and used after retrieval to guide the generation of a realistic mistake.

```python
chunks = []

for _, row in df.iterrows():
    
    chunk = {
        "text": row["question"],

        "metadata": {
            "qid": row["qid"],
            "student_incorrect_solution": row["student_incorrect_solution"],
            "teacher_described_confusion": row["teacher_described_confusion"],
            "ground_truth": row["ground_truth"],
           
        }
    }
    
    chunks.append(chunk)

len(chunks)
```

# Embedding

### Models selection

We use open-source models for both embeddings and generation to keep the pipeline fully local and reproducible.

For embeddings, we use all-MiniLM-L6-v2, a lightweight model that provides efficient semantic similarity while requiring low computational resources. Since the questions are relatively short (max 148 words), the model’s context capacity is sufficient while maintaining fast inference.

For the generator (used later for Agent 1 and Agent 2), we plan to use Mistral 7B through Ollama, which is an open-source LLM that offers good performance while remaining feasible to run locally.

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
texts = [chunk["text"] for chunk in chunks]

embeddings = model.encode(
    texts,
    show_progress_bar=True,
    normalize_embeddings=True
)

len(embeddings), len(embeddings[0])
```

## ChromaDB

```python
import chromadb

# create database (simple version)
client = chromadb.Client()

# create collection
collection = client.get_or_create_collection(name="mathdial")

# prepare data
documents = [chunk["text"] for chunk in chunks]
metadatas = [chunk["metadata"] for chunk in chunks]
ids = [str(i) for i in range(len(chunks))]

# store embeddings
collection.add(
    documents=documents,
    embeddings=embeddings,
    metadatas=metadatas,
    ids=ids
)
```

```python
collection.count()
```

```python
def retrieve_examples(query, collection, top_k=5):
    
    # convert query into embedding
    query_embedding = model.encode(query).tolist()
    
    # search similar questions
    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=top_k
    )
    
    retrieved_questions = results["documents"][0]
    retrieved_metadata = results["metadatas"][0]
    
    examples = []
    
    for i in range(len(retrieved_questions)):
        
        example = {
            "question": retrieved_questions[i],
            "qid": retrieved_metadata[i]["qid"],
            "student_incorrect_solution": retrieved_metadata[i]["student_incorrect_solution"],
            "teacher_described_confusion": retrieved_metadata[i]["teacher_described_confusion"],
            "ground_truth": retrieved_metadata[i]["ground_truth"],
            
        }
        
        examples.append(example)
    
    return examples
```

```python
query = "Nancy is filling an aquarium for her fish"

example = retrieve_examples(query, collection, top_k=3)

example
```

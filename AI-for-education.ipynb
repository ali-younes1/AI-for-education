## Text preprocessing

No classical NLP preprocessing is applied, such as:

- removing stop words
- stemming
- lemmatization
- manual tokenization
- lowercasing everything
- removing punctuation

Since the RAG pipeline relies on semantic embeddings, the model already handles tokenization and preserves meaning. Modifying the text could reduce important context, especially in tutoring dialogues.

Therefore, the text is kept in its original form.

# Chunking

### Chunking strategy for Agent 1

For Agent 1, retrieval is based only on the question. At the beginning of the interaction, no incorrect student solution exists yet, since Agent 1 must first generate a flawed response.

Therefore, each chunk is indexed using the question only, while fields such as student_incorrect_solution and teacher_described_confusion are kept as metadata and used after retrieval to guide the generation of a realistic mistake.

```python
chunks = []

for _, row in df.iterrows():

    chunk = {
        "text": row["question"],

        "metadata": {
            "qid": row["qid"],
            "student_incorrect_solution": row["student_incorrect_solution"],
            "teacher_described_confusion": row["teacher_described_confusion"],
            "ground_truth": row["ground_truth"],

        }
    }

    chunks.append(chunk)

len(chunks)
```

# Embedding

### Models selection

We use open-source models for both embeddings and generation to keep the pipeline fully local and reproducible.

For embeddings, we use all-MiniLM-L6-v2, a lightweight model that provides efficient semantic similarity while requiring low computational resources. Since the questions are relatively short (max 148 words), the model’s context capacity is sufficient while maintaining fast inference.

We use Qwen2.5-Instruct as the generator (used later for Agent 1 and Agent 2) because instruction-tuned models are optimized to follow prompts and produce structured, step-by-step explanations, which is well suited for RAG and tutoring tasks.

```python
import chromadb

# create database (simple version)
client = chromadb.Client()

# create collection
collection = client.create_collection(name="mathdial", configuration={"hnsw": {"space": "cosine"}})

# prepare data
documents = [chunk["text"] for chunk in chunks]
metadatas = [chunk["metadata"] for chunk in chunks]
ids = [str(i) for i in range(len(chunks))]

# store embeddings
collection.add(
    documents=documents,
    embeddings=embeddings,
    metadatas=metadatas,
    ids=ids
)
```

```python
def retrieve_examples(query, collection, top_k=5):

    # convert query into embedding
    query_embedding = embedding_model.encode(query).tolist()

    # search similar questions
    results = collection.query(
        query_embeddings=[query_embedding],

        n_results=top_k
    )

    retrieved_questions = results["documents"][0]
    retrieved_metadata = results["metadatas"][0]

    examples = []

    for i in range(len(retrieved_questions)):

        example = {
            "question": retrieved_questions[i],
            "qid": retrieved_metadata[i]["qid"],
            "student_incorrect_solution": retrieved_metadata[i]["student_incorrect_solution"],
            "teacher_described_confusion": retrieved_metadata[i]["teacher_described_confusion"],
            "ground_truth": retrieved_metadata[i]["ground_truth"],

        }

        examples.append(example)

    return examples
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
# Pytorch software library used to run neural networks.
import torch

model_name = "Qwen/Qwen2.5-3B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    # 1.5B parameters × 2 bytes ≈ 3 GB VRAM
    torch_dtype=torch.float16,
    # automatically uses GPU if available, otherwise CPU
    device_map="auto",


)
```

```python
import json

def generate_text(messages, max_new_tokens=200):
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    model_inputs = tokenizer([text], return_tensors="pt").to(model.device)

    generated_ids = model.generate(
        **model_inputs,
        max_new_tokens=max_new_tokens,

    )

    generated_ids = [
        output_ids[len(input_ids):]
        for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
    ]

    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
    return response.strip()
```

```python
def is_math_problem(problem):
    messages = [
        {
            "role": "system",
            "content": """You are a classifier.

Return ONLY valid JSON in one of these two formats:
{"is_math_problem": true}
{"is_math_problem": false}

No extra text."""
        },
        {
            "role": "user",
            "content": problem
        }
    ]

    response = generate_text(messages, max_new_tokens=30)

    try:
        data = json.loads(response)
        return data.get("is_math_problem", False)
    except:
        return False
```

```python
def run_agent2(problem, retrieved_example, history_text):
    messages = [
        {
            "role": "system",
            "content": """You are an evaluator.

Your role is to compare the student's reasoning with the ground truth and decide what Agent 1 should do next.

Instructions:
- Use the retrieved similar problem and ground truth as reference
- Check whether the student correctly understands the key idea
- Identify what is still missing or incorrect
- Give one short instruction for Agent 1
- Keep the output concise

Return ONLY valid JSON in this format:
{
  "student_understanding": "none/partial/good",
  "missing_concept": "short text",
  "agent1_instruction": "short instruction"
}"""
        },
        {
            "role": "user",
            "content": f"""Real student problem:
{problem}

Retrieved similar problem:
{retrieved_example['question']}

Retrieved ground truth:
{retrieved_example['ground_truth']}

Conversation history:
{history_text}"""
        }
    ]

    response = generate_text(messages, max_new_tokens=150)

    try:
        return json.loads(response)
    except:
        return {
            "student_understanding": "unknown",
            "missing_concept": "Could not parse output.",
            "agent1_instruction": "Keep the misunderstanding and ask the student to explain their reasoning."
        }
```

```python
def run_agent1(problem, retrieved_example, history_text, agent2_instruction):
    messages = [
        {
            "role": "system",
            "content": """You are a flawed problem-solving AI.

Your role is to give an incorrect or partially incorrect response to the problem, based on the retrieved example and the conversation history.

Instructions:
- Start with a realistic misunderstanding
- Use the retrieved incorrect solution and confusion as guidance
- Do not reveal the correct final answer too early
- Respond to the student's latest message
- Revise your reasoning only if the student gives a convincing correction
- Keep the response short, natural, and believable
- Do not mention the retrieved example or hidden evaluation

Return only the reply text."""
        },
        {
            "role": "user",
            "content": f"""Real student problem:
{problem}

Retrieved similar problem:
{retrieved_example['question']}

Retrieved incorrect solution:
{retrieved_example['student_incorrect_solution']}

Retrieved confusion:
{retrieved_example['teacher_described_confusion']}

Conversation history:
{history_text}

Controller instruction:
{agent2_instruction}"""
        }
    ]

    return generate_text(messages, max_new_tokens=120)
```

```python
def format_history(history):
    return "\n".join(f"{m['role']}: {m['content']}" for m in history)
```

```python
def run_pipeline(problem, collection, conversation_history):
    if not is_math_problem(problem):
        return {
            "is_math_problem": False,
            "retrieved_example": None,
            "agent2_result": None,
            "agent1_reply": "This does not look like a math problem."
        }

    retrieved_examples = retrieve_examples(problem, collection, top_k=1)

    if not retrieved_examples:
        return {
            "is_math_problem": True,
            "retrieved_example": None,
            "agent2_result": None,
            "agent1_reply": "No similar example was retrieved."
        }

    best_example = retrieved_examples[0]
    history_text = format_history(conversation_history)

    agent2_result = run_agent2(
        problem=problem,
        retrieved_example=best_example,
        history_text=history_text
    )

    agent1_reply = run_agent1(
        problem=problem,
        retrieved_example=best_example,
        history_text=history_text,
        agent2_instruction=agent2_result["agent1_instruction"]
    )

    return {
        "is_math_problem": True,
        "retrieved_example": best_example,
        "agent2_result": agent2_result,
        "agent1_reply": agent1_reply
    }
```

```python
problem = "Nancy is filling an aquarium for her fish. She fills it halfway and goes to answer the door. While she's gone, her cat knocks the aquarium over and spills half the water in it. Then Nancy comes back and triples the amount of water in the aquarium. If the aquarium is 4 feet long, 6 feet wide, and 3 feet high, how many cubic feet of water are in the aquarium?"

conversation_history = [
    {
        "role": "student",
        "content": "I think the answer is 54 ."
    }
]

result = run_pipeline(problem, collection, conversation_history)

print("is_math_problem:", result["is_math_problem"])

print("\nRetrieved example:")
print(result["retrieved_example"])

print("\nAgent 2 result:")
print(result["agent2_result"])

print("\nAgent 1 reply:")
print(result["agent1_reply"])
```

```python
problem = "Nancy is filling an aquarium for her fish. She fills it halfway and goes to answer the door. While she's gone, her cat knocks the aquarium over and spills half the water in it. Then Nancy comes back and triples the amount of water in the aquarium. If the aquarium is 4 feet long, 6 feet wide, and 3 feet high, how many cubic feet of water are in the aquarium?"

conversation_history = [
    {
        "role": "student",
        "content": "I think the answer is 54 ."
    }
]

# first run
result = run_pipeline(problem, collection, conversation_history)

print("Agent 1 first reply:")
print(result["agent1_reply"])

# save Agent 1 reply into history
conversation_history.append({
    "role": "assistant",
    "content": result["agent1_reply"]
})

# student argues back
conversation_history.append({
    "role": "student",
    "content": "I do not agree with the part where you add 18 and 54. If Nancy triples the remaining 18 cubic feet, doesn't that already make the final amount 54?"
})

# second run
result2 = run_pipeline(problem, collection, conversation_history)


print("\nAgent 1 second reply:")
print(result2["agent1_reply"])
print("\nAgent 2 result:")
print(result["agent2_result"])
```
